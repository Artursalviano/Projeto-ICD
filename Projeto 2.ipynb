```python
import pandas as pd
import numpy as np

relatorio_mobilidadeBR = pd.read_csv('2020_BR_Region_Mobility_Report.csv')
relatorio_mobilidadeBR
```

Utilizando Apenas os dados de mobilidade da Paraíba pois para calcular as medidadas de centralidade e dispersão dos dados do país inteiro é muito demorado

```python
relatorio_mobilidade_estados =relatorio_mobilidadeBR.set_index('sub_region_1')
relatorio_mobilidade_paraiba = relatorio_mobilidade_estados.loc['State of Paraíba']
relatorio_mobilidade_paraiba
```

```python
print('centralidades\n\n')
print('Médias\n\n',relatorio_mobilidade_paraiba.mean())
print('\n Medianas \n\n',relatorio_mobilidade_paraiba.median())
```

```python
print('Dispersão \n')
print('Variancia\n\n',relatorio_mobilidade_paraiba.var())
std = relatorio_mobilidade_paraiba.describe()
print('\n Desvio padrão \n\n',std.loc['std'])
```

```python
print('boxplot\n')
relatorio_mobilidade_paraiba.boxplot(column=['retail_and_recreation_percent_change_from_baseline','grocery_and_pharmacy_percent_change_from_baseline'
                                            ,'parks_percent_change_from_baseline','transit_stations_percent_change_from_baseline'
                                            , 'workplaces_percent_change_from_baseline', 'residential_percent_change_from_baseline' ], rot=90, fontsize = 8)
```
